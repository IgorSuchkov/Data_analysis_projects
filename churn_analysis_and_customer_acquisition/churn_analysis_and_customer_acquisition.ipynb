Получается, если минимальное значение userid домножить на 2, то получим уникальные значения для дубликатов.

```python
for i in data_rare[data_rare.duplicated(subset=['userid']) == True].index:
    data_rare['userid'][i] *= 2
```

# Прогнозирования оттока клиентов

## Модель прогнозирования оттока клиентов

Построим модель бинарной классификации пользователей, где целевой признак — факт оттока клиента

- Разобьём данные на обучающую и валидационную выборку функцией train_test_split().
- Обучим модель на train-выборке двумя способами:
 1. логистической регрессией
 2. случайным лесом

```python
df['city'] = df['city'].replace('ярославль','1').replace('ростов','2').replace('рыбинск','3').astype('int')
df['gender'] = df['gender'].replace('ж','0').replace('м','1').astype('int')

df = df[['score', 'age', 'equity', 'balance', 'products', 'credit_card', 'last_activity', 'est_salary',
        'churn', 'city', 'gender']]

df = df.fillna(0) # заполним пропущенные значения на 0, чтобы не мешать обучению
```

```python
# разделим наши данные на признаки (матрица X) и целевую переменную (y)
X = df.drop(columns = ['churn']) 
y = df['churn']
```

```python
# разделяем модель на обучающую и валидационную выборки
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
# обучим StandartScaler на обучающей выборке
scaler = StandardScaler()
scaler.fit(X_train)
# Преобразуем обучающие и валидационные наборы данных
X_train_st = scaler.transform(X_train)
X_test_st = scaler.transform(X_test)
```

```python
# зададим алгоритм для модели логистической регрессии
lr_model = LogisticRegression()
# обучим модель
lr_model.fit(X_train_st, y_train)
# воспользуемся уже обученной моделью
lr_predictions = lr_model.predict(X_test_st)
lr_probabilities = lr_model.predict_proba(X_test_st)[:,1]
```

```python
# создадим датафрейм с признаками и их весами
lr_features = pd.DataFrame({'features' : X_train.columns, 'coeff' : lr_model.coef_.ravel()}) # воспользуйтесь методом coef_
# обратим коэф-ты по модулю для оценки влияния на признаки 
lr_features['coeff_abs'] = abs(lr_features['coeff']) 
# выведите упорядоченный по модулю коэффициентов датафрейм с признаками
lr_features.sort_values('coeff_abs', ascending = False).reset_index(drop=True)
```

Самые важные признаки - last_activity, products, gender, balance, equity.

```python
# зададим алгоритм для новой модели на основе алгоритма случайного леса
rf_model = RandomForestClassifier(n_estimators = 100, random_state = 0)
# обучим модель случайного леса
rf_model.fit(X_train_st, y_train)
# воспользуемся уже обученной моделью, чтобы сделать прогнозы
rf_predictions = rf_model.predict(X_test_st)
rf_probabilities = rf_model.predict_proba(X_test_st)[:,1]
```

```python
# создадим датафрейм с признаками и их весами
rf_features = pd.DataFrame({'feature' : X_train.columns, 'importance' : rf_model.feature_importances_}) # методом coef_

# выведите упорядоченный по модулю коэффициентов датафрейм с признаками
rf_features.sort_values('importance', ascending = False).reset_index(drop=True)
```

Самые важные признаки - score, balance, est_salary, age.

```python
# определим функцию, которая будет выводить наши метрики
def print_metrics(y_true, y_pred, y_proba, title = 'Метрики классификации'):
    print(title)
    print('\tAccuracy: {:.2f}'.format(accuracy_score(y_true, y_pred)))
    print('\tPrecision: {:.2f}'.format(precision_score(y_true, y_pred)))
    print('\tRecall: {:.2f}'.format(recall_score(y_true, y_pred)))
```

```python
# посчитаем все метрики
print_metrics(y_test, lr_predictions, lr_probabilities, title = 'Метрики для модели логистической регрессии:')
print_metrics(y_test, rf_predictions, rf_probabilities, title = 'Метрики для модели случайного леса:')
```

У случайного леса метрики лучше. Логистическая регрессия ведёт себя хуже.

## Кластеризацию пользователей

Стандартизируем данные и построим матрицу расстояний функцией linkage() на стандартизованной матрице признаков и нарисуем дендрограмму. На основании полученного графика предположим, какое количество кластеров можно выделить.

```python
from sklearn.cluster import KMeans
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
# стандартизируем данные
sc = StandardScaler() # создаём объект класса scaler (нормализатор)
# обучаем нормализатор и преобразуем набор данных без учёта оттока, чтобы это не повлияло на модель
X_sc = sc.fit_transform(df.drop(columns = 'churn')) 
```

```python
# построим матрицу расстояний
linked = linkage(X_sc, method = 'ward')
```

```python
# визуализируем таблицу расстояний linked
plt.figure(figsize=(15, 10))  
dendrogram(linked, orientation='top', truncate_mode='lastp')
plt.title('Иерархия кластеров для клиентов банка');
```

Получилось 3 кластера.

## Обучим модель кластеризации на основании алгоритма K-Means и спрогнозируем кластеры клиентов. Число кластеров принимаем n=3.

```python
# задаём модель k_means с числом кластеров 3
km = KMeans(n_clusters = 3, random_state = 0) # 
# прогнозируем кластеры для наблюдений
labels = km.fit_predict(X_sc)

# сохраняем метки кластера в поле нашего датасета
df['cluster_km'] = labels
```

```python
# выводим статистику по средним значениям наших признаков по кластеру
df.groupby(['cluster_km']).mean().T
```

Количество записей в каждом кластере.

```python
df.groupby('cluster_km')['gender'].count()
```

Кластеры получились, можно сказать, схожи по количеству.

Оценим качетство проведенной кластеризации

```python
print('Silhouette_score: = ', round(silhouette_score(X_sc, labels), 3) * 100)
```

Получаем Silhouette_score на уровне около 13 процентов. Что ниже среднего показателя.

# Проверка гипотез

## Равенство доходов

Сформулируем первую гипотезу.
- H0:Средний доход отточных и неотточных клиентов равен
- H1:Средний доход отточных и неотточных клиентов различается
 
α=5% критический уровень статистической значимости.

Разделим данные по оттоку и проверим значения дисперсий в получившихся совокупностях.

```python
print (stat.variance(df['est_salary']))
```

```python
print (stat.variance(df.query('churn == 1')['est_salary']))
print (stat.variance(df.query('churn == 0')['est_salary']))
```

Необходимо определиться со статистическим методом.

Для того, чтобы определить тест, необходимо определить нормально ли у нас распределены данные в выборках? И есть ли выбросы?

С выбросами мы поработали на этапе обработке данных.

```python
print(shapiro(df.query('churn == 1')['est_salary']), 'Шапиро для отточных')
print(shapiro(df.query('churn == 0')['est_salary']), 'Шапиро для оставшихся', '\n')

print(skew(df.query('churn == 1')['est_salary'], axis=0, bias=True), 'ассиметрия для отточных')
print(skew(df.query('churn == 0')['est_salary'], axis=0, bias=True), 'ассиметрия для оставшихся' '\n')

print(kurtosis(df.query('churn == 1')['est_salary'], axis=0, bias=True), 'эксцесс для отточных')
print(skew(df.query('churn == 0')['est_salary'], axis=0, bias=True), 'эксцесс для оставшихся')
```

- По результат теста Шапиро говорит, на о том, что данные выборки взяты из не нормально распредленной ГС.(alpha = 0.05)
- Результат теста на ассиметрию далек от нормального распределение.
- Результат тест на разброс значений, также далек от нормального распределения.

Вывод t-тест — нельзя применить, так как значения не распределены нормально. Применяем тест Манн-Уитни.

equal_var=False - так как выборки миеют разную дисперсию и размерность.

```python
df_churn = df.query('churn == 1')['est_salary']
df_not_churn = df.query('churn == 0')['est_salary']

alpha = 0.05
results = st.ttest_ind(
    df_churn, 
    df_not_churn,
    equal_var=False)

print('p-значение:', results.pvalue)

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

```python
df_churn = df.query('churn == 1')['est_salary']
df_not_churn = df.query('churn == 0')['est_salary']

alpha = 0.05  # критический уровень статистической значимости

results = st.mannwhitneyu(df_churn, df_not_churn)

print('p-значение: ', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу: разница статистически значима')
else:
    print(
        'Не получилось отвергнуть нулевую гипотезу, вывод о различии сделать нельзя'
    ) 
```

Отвергаем нулева гипотезу, считаем, что средний доход отточных и неотточных клиентов разлицаются.

Критерий Манна-Уитни менее чувствительный к выбросам и ему не нужно условие нормальности распределения. Поэтому берем его результат.

## Равенство возраста

Сформулируем вторую гипотезу
- H0:Средний возраст отточных и неотточных клиентов равен
- H1:Средний возраст отточных и неотточных клиентов различается
 
α=5% критический уровень статистической значимости.

```python
# проверим значения дисперсий
print (stat.variance(df.query('churn == 1')['age']))
print (stat.variance(df.query('churn == 0')['age']))
```

```python
print(shapiro(df.query('churn == 1')['age']), 'Шапиро для отточных')
print(shapiro(df.query('churn == 0')['age']), 'Шапиро для оставшихся', '\n')

print(skew(df.query('churn == 1')['age'], axis=0, bias=True), 'ассиметрия для отточных')
print(skew(df.query('churn == 0')['age'], axis=0, bias=True), 'ассиметрия для оставшихся' '\n')

print(kurtosis(df.query('churn == 1')['age'], axis=0, bias=True), 'эксцесс для отточных')
print(skew(df.query('churn == 0')['age'], axis=0, bias=True), 'эксцесс для оставшихся')
```

- По результат теста Шапиро говорит, на о том, что данные выборки взяты из не нормально распредленной ГС.(alpha = 0.05)
- Результат теста на ассиметрию далек от нормального распределение.
- Результат тест на разброс значений, также далек от нормального распределения.

Вывод t-тест — нельзя применить, так как значения не распределены нормально. Применяем тест Манн-Уитни.

```python
df_churn = df.query('churn == 1')['age']
df_not_churn = df.query('churn == 0')['age']

alpha = 0.05  # критический уровень статистической значимости

results = st.mannwhitneyu(df_churn, df_not_churn)

print('p-значение: ', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу: разница статистически значима')
else:
    print(
        'Не получилось отвергнуть нулевую гипотезу, вывод о различии сделать нельзя'
    ) 
```

Есть основания отвергнуть нулевую гипотезу, считаем, что средний возраст отточных и неотточных клиентов различается.

на данном шаге проверены две гипотезы:
- о равенстве среднего дохода отточных и не отточных клиентов;
- о равенстве среднего возраста отточных и не отточных клиентов.

по итогу можем сказать:
- средний доход отточных и неотточных клиентов различается;
- средний возраст отточных и неотточных клиентов различается
